I want Neha and Dhruv create two data frames
 but make sure one column should be common in them


```python
ms1 ={"StudentName":["Dhru","Gaurav"],"Distance":[40,50,]}
ms2 ={"StudentName":["Ashish","Gaurav"],"Lucky":[6,5]}
```

```python
ms1=pd.DataFrame(ms1)
ms1
```

```python
ms2=pd.DataFrame(ms2)
ms2
```

```python
# Inner Join
msa=pd.merge(ms1,ms2,on='StudentName',how='inner')
```

```python
msa
```

```python
#Left Join
msb=pd.merge(ms1,ms2,on='StudentName',how='left')
```

```python
msb
```

```python
#Right Join
msc=pd.merge(ms1,ms2,on='StudentName',how='right')
msc
```

```python
#Outer Join
msd=pd.merge(ms1,ms2,on='StudentName',how='outer')
msd
```
